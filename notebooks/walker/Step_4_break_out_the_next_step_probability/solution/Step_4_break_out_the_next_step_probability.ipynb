# Simulate a trajectory

## Using a Gaussian next step proposal

```python
context_map = hills_context_map(size=200)
walker = Walker(
    context_map=context_map, 
    next_step_proposal=gaussian_next_step_proposal,
    next_step_proposal_arguments={'sigma_i': 3, 'sigma_j': 4}
)

# Sample a next step 1000 times
i, j = 100, 50
trajectory = []
for _ in range(1000):
    i, j = walker.sample_next_step(i, j)
    trajectory.append((i, j))
```

```python
plot_trajectory(trajectory, walker.context_map)
```
